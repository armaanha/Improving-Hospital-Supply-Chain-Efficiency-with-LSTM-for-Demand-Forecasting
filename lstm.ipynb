```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import TimeSeriesSplit
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, BatchNormalization, Embedding
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, TensorBoard
import datetime
import os

# Load the dataset
data = pd.read_csv('/Users/armaanhaque/Documents/ML project/dataset/inventory_data.csv')

# Preprocess the dataset
data['Date'] = pd.to_datetime(data['Date'])
data = data.set_index('Date')
data = data.asfreq('D')

# Feature Engineering
data['day_of_week'] = data.index.dayofweek
data['month'] = data.index.month

data_avg_usage = data.groupby(['Item_Name']).resample('D').agg({'Avg_Usage_Per_Day': 'mean',
                                                                'Current_Stock': 'mean',
                                                                'Min_Required': 'mean',
                                                                'Max_Capacity': 'mean',
                                                                'day_of_week': 'first',
                                                                'month': 'first'}).reset_index()
data_avg_usage = data_avg_usage.set_index('Date')

# Remove 'X-ray Machine' from consideration
data_avg_usage = data_avg_usage[data_avg_usage['Item_Name'] != 'X-ray Machine']

# Create logs directory if it doesn't exist
if not os.path.exists('logs'):
    os.makedirs('logs')

# LSTM model for each item
rmse_values = {}  # To store RMSE values for each item
for item_name in data_avg_usage['Item_Name'].unique():
    item_data = data_avg_usage[data_avg_usage['Item_Name'] == item_name]
    features = ['Avg_Usage_Per_Day', 'Current_Stock', 'Min_Required', 'Max_Capacity', 'day_of_week', 'month']
    item_data = item_data[features]

    # Fill NaN values using linear interpolation
    item_data = item_data.interpolate(method='linear')

    # Normalize the data using only training data to fit the scaler
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(item_data)

    # Split data using TimeSeriesSplit with larger test size
    tscv = TimeSeriesSplit(n_splits=3)
    train_data, test_data = None, None
    for train_index, test_index in tscv.split(scaled_data):
        train_data, test_data = scaled_data[train_index], scaled_data[test_index]

    # Function to create sequences
    def create_sequences(data, seq_length):
        X, y = [], []
        for i in range(len(data) - seq_length):
            X.append(data[i:i + seq_length, :])
            y.append(data[i + seq_length, 0])  # Predicting 'Avg_Usage_Per_Day'
        return np.array(X), np.array(y)

    # Sequence length of 60 days
    sequence_length = 60

    # Create sequences for training and testing
    X_train, y_train = create_sequences(train_data, sequence_length)
    X_test, y_test = create_sequences(test_data, sequence_length)

    # Reshape input to be [samples, time steps, features]
    X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], X_train.shape[2]))
    X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], X_test.shape[2]))

    # Define the LSTM model with added features
    model = Sequential()
    model.add(LSTM(units=80, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
    model.add(Dropout(0.4))
    model.add(BatchNormalization())
    model.add(LSTM(units=30, return_sequences=False))
    model.add(Dropout(0.4))
    model.add(Dense(units=1))

    # Compile the model
    model.compile(optimizer='adam', loss='mean_squared_error')

    # TensorBoard callback for visualization
    log_dir = f"logs/fit/{item_name}_" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
    tensorboard_callback = TensorBoard(log_dir=log_dir, histogram_freq=1)

    # Add early stopping and model checkpoint callbacks
    early_stopping = EarlyStopping(monitor='val_loss', patience=15, restore_best_weights=True)
    model_checkpoint = ModelCheckpoint(f'best_model_{item_name}.h5', save_best_only=True, monitor='val_loss', mode='min')

    # Train the model
    model.fit(X_train, y_train, epochs=100, batch_size=16, validation_data=(X_test, y_test),
              callbacks=[early_stopping, model_checkpoint, tensorboard_callback], verbose=2)

    # Make predictions using the best model
    predictions = model.predict(X_test)
    predictions = scaler.inverse_transform(np.concatenate([predictions, np.zeros((len(predictions), X_test.shape[2] - 1))], axis=1))[:, 0]

    # Prepare actual values for comparison
    actual_train_values = scaler.inverse_transform(train_data[sequence_length:])[:, 0]
    actual_test_values = scaler.inverse_transform(test_data[sequence_length:])[:, 0]

    # Plotting actual vs predicted values
    plt.figure(figsize=(12, 6))
    # Plotting Training Data
    plt.plot(range(sequence_length, sequence_length + len(actual_train_values)), actual_train_values, color='blue', label='Training Data')
    # Plotting Test Data
    test_start_idx = sequence_length + len(actual_train_values)
    plt.plot(range(test_start_idx, test_start_idx + len(actual_test_values)), actual_test_values, color='red', label='Test Data')
    # Plotting Predicted Values
    plt.plot(range(test_start_idx, test_start_idx + len(predictions)), predictions, color='green', linestyle='--', label='Predicted Test Data')
    plt.title(f'Actual vs Predicted Avg Usage Per Day for Item_Name: {item_name}')
    plt.xlabel('Days')
    plt.ylabel('Avg Usage Per Day')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Evaluation Metrics (RMSE for each item)
    from sklearn.metrics import mean_squared_error
    rmse = np.sqrt(mean_squared_error(actual_test_values, predictions))
    rmse_values[item_name] = rmse
    print(f'RMSE for {item_name}: {rmse}')

# Print RMSE for all items
print("\nRMSE Values for All Items:")
for item, value in rmse_values.items():
    print(f'{item}: {value}')


```

```python
from statsmodels.tsa.stattools import adfuller

result = adfuller(data['Avg_Usage_Per_Day'])
print(f'p-value: {result[1]}')
if result[1] < 0.05:
    print("The data is stationary.")
else:
    print("The data is not stationary.")
```
